# Import Dependancy

```python
import pandas as pd
```

# Load Clean Datasets

```python
Currys=pd.read_csv("../Clean_Datasets/All_Currys_Products.csv")
Very=pd.read_csv("../Clean_Datasets/All_Very_Products.csv")
```

```python
# View Currys Dataset

Currys
```

# Merge Datasets

```python
# Join both datasets

Merged=pd.merge(Currys, Very, how='inner', left_on = 'product_name', right_on = 'product_name')
```

```python
Merged.rename(columns={'category_x':'Category', 'price_x':"Curry_Price", 'customer_rating_x':"Curry_Rating",
       'rating_count_x':"Curry_Rating_Count", 'price_y':"Very_Price", 'customer_rating_y':"Very_Rating",
       'rating_count_y':"Very_Rating_Count"},inplace=True)
del Merged["category_y"]


Merged["Price_Difference"]=((Merged["Curry_Price"]-Merged["Very_Price"])/Merged["Curry_Price"])*100
Merged["Rating_Difference"]=((Merged["Curry_Rating"]-Merged["Very_Rating"])/5)*100
```

```python
# View merged dataset

Merged
```

# Analysis of Merged Data

```python
Merged["Rating_Difference"]=((Merged["Curry_Rating"]-Merged["Very_Rating"])/5)*100
```

```python
# Average price difference

Merged["Price_Difference"].mean()
```

```python
# Average customer ratings difference

Merged["Rating_Difference"].mean()
```

```python
Filtered=Merged.loc[Merged["Curry_Rating_Count"]>=50]
Filtered=Filtered.loc[Filtered["Very_Rating_Count"]>=50]
```

```python
Filtered["Rating_Difference"].mean()
```

```python
# Save merged and filtered datasets to CSV

Merged.to_csv("Merged.csv")
Filtered.to_csv("Filtered.csv")
```
